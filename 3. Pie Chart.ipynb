```python
import matplotlib.pyplot as plt
```

```python
digit = [60, 40, 10, 50, 45]
name = ['sixty' , 'fourty', 'ten', 'fivty', 'fourtyfive']

plt.pie(digit , labels =name, explode = [0.1,0,0,0,0], autopct = '%1.f%%')   # Explode Prameter is optional

plt.title('Pie chart')

# plt.legend()                       # Legend is optional
# plt.tight_layout()

plt.show()
```
