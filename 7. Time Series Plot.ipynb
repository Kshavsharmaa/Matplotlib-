### Time Series Plot 

```python
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates,y, linestyle = '-', label = 'Date')

plt.gcf().autofmt_xdate()

plt.title('Date Plot')
plt.xlabel('Dates')
plt.ylabel('No. of Dates')

plt.legend()
plt.grid()

plt.tight_layout()

plt.show()
```
