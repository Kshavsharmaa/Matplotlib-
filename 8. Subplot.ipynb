### Subplots

  - With the subplot() function you can draw multiple plots in one figure:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")

plt.tight_layout()

plt.show()
```
