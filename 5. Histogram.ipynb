### Histogram 

```python
import matplotlib.pyplot as plt
```

```python
ages = [18,19,21,25,26,26,30,32,38,45,55]
bins = [10,20,30,40,50,60]

plt.hist(ages, bins = bins, edgecolor = 'red', label = 'Ages Ranges')

plt.title('Histogram')
plt.xlabel('Ages')
plt.ylabel('Ages Range')
plt.legend()

plt.tight_layout()

plt.show()
```
